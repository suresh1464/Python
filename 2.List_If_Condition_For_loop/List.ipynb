Lists allow you to store sequential data
Lists are ordered, meaning each item has a fixed position unless explicitly changed
Python lists can hold different data types in a single list, including numbers,strings, and other list. This means they are heterogeneous. Example my_list = ["Car",4.5,True]
Lists are mutable , allowing for elements to be added,removed and changed within the same list
Lists slicing lets you access a specific range of elements quickly,using the syntax list [start:end:step]

```python
items = ["bread","pasta","fruits","veggies"]
items
```

```python
items[0]
```

```python
items.append("butter")
items
```

```python
items.remove("butter")
items
```

```python
items[1] = "Cheese"  # Strings are immutable where as Lists are mutable
items
```

```python
items[3:] = ["almonds"]
items
```
